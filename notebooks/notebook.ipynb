```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/train_split.csv")
texts = (df["Title"] + " " + df["Description"]).astype(str)

word_lengths = texts.apply(lambda x: len(x.split()))

print("Средняя длина (слов):", np.mean(word_lengths))
print("p90:", np.percentile(word_lengths, 90))
print("p99:", np.percentile(word_lengths, 99))
```

```python
char_lengths = texts.apply(len)

print("Средняя длина (символов):", np.mean(char_lengths))
print("p90:", np.percentile(char_lengths, 90))
print("p99:", np.percentile(char_lengths, 99))
```

```python
from collections import Counter
import re

def tokenize_words(text):
    return re.findall(r"\w+", text.lower())

counter = Counter()

for t in texts[:50000]:  # достаточно подвыборки
    counter.update(tokenize_words(t))

print("Уникальных слов:", len(counter))
```

```python
rare = sum(1 for w, c in counter.items() if c == 1)
print("Редкие слова (%):", round(rare / len(counter) * 100, 2))
```
